# Plot Marker, Linestyle & Color

```python
ypoints = np.array([5,8,1,10,15])

plt.plot(ypoints, marker="d",c="red",linestyle ="dashdot",ms="15",mec="blue",mfc="green")
plt.xlabel("X-Axis")
plt.ylabel("y-Axis")
plt.title("Line Plot")
plt.show
```

# Text Annotation

```python
#draw plot as figure
fig, ax=plt.subplots()
x= np.linspace(0,20,1000)
y= np.cos(x)
ax.plot(x,y)
ax.axis('equal')
ax.annotate("local maximum",xy=(2*np.pi,1),xytext=(10,4),arrowprops=dict(facecolor='red'))
ax.annotate("local minimum",xy=(5*np.pi,-1),xytext=(10,-4),arrowprops=dict(facecolor='green'))
plt.show()
```

# 3D Plot

```python
#import mplot3d toolkit,numpy,matplotlib

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()

#3d projection
ax=plt.axes(projection='3d')
z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

ax.plot3D(x,y,z,'green')
ax.set_title('3Dlineplot')
plt.show()
```

# Pie Chart

```python
#creating data set

country =['India','China','Pakistan','USA','Russia']
population=[131,135,95,40,50]
explode = (0,0,0.1,0,0.3,)

plt.pie(population,labels = country,explode = explode,shadow = True,autopct ='%1.1f%%')
#fig=plt.figure(figsize=(30,80))
plt.savefig('population_graph.jpeg',format="jpeg",dpi=200)
plt.show()
```
